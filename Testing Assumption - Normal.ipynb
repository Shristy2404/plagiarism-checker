# Testing Normal Distribution assumption

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('final_dataset.csv')
df.head()
```

```python
def test_normality(arr):
    sns.distplot(arr)
```

```python
unique_auths = df['author'].unique()
for auth in unique_auths:
    df_sub = df[df['author']==auth]
    test_normality(df_sub['N2'])
plt.show()
```
